### BLEU SCORE

Esta métrica surgiu com o objetivo de avaliar traduções de texto geradas por máquinas, comparando com o mesmo texto gerado por um ser humano. 
A idéia do algoritmo é avaliar um conjunto de sentenças, pontuando a frase com um valor entre 0 e 1, onde 1 é uma similaridade perfeita entre o texto gerado e o texto de teste. Recupera se então a média dos valores das diferentes sentenças e temos um score.
Existem diversas bibliotecas que implementam esse algoritmo, como nltk, evaluate e torchmetrics, que exemplifico logo abaixo.


```python
#! pip install torchmetrics
```

```python
from torchmetrics.text import BLEUScore

bleu = BLEUScore()
candidates = ["este é um exemplo"]                  # valor a ser testado
references = [["este é um exemplo"]]                # valor de referência

score = bleu(candidates, references)
print(f'Score BLEU: {score:.2f}')
```

```python
bleu = BLEUScore()
candidates = ["este é um exemplo "]                  # valor a ser testado
references = [["este é um exemplo diferente"]]       # valor de referência

score = bleu(candidates, references)
print(f'Score BLEU: {score:.2f}')
```

```python
bleu = BLEUScore()
# Lista de frases candidatas (como strings)
candidatos = [
    "este é um exemplo",
    "o gato está no telhado",
    "a máquina está funcionando"
]

# Lista de listas de frases de referência (como strings)
referencias = [
    ["este é um exemplo"],
    ["o gato não está no telhado"],
    ["a máquina não está funcionando"]
]

score = bleu(candidatos, referencias)
print(f'Score BLEU: {score:.2f}')
```

O paper original pode ser consultado em https://aclanthology.org/P02-1040/
Reparem que este método depende fortemente da tokenização e é mais utilizado para avaliar traduções. Pode ser usado para avaliar LLMs desde que se testem sentenças curtas, traduções e pequenos resumos.
